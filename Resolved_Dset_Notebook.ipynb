```python

import glob, os
import os.path as op
import pandas as pd
import numpy as np
```

```python
#Get subject number
dset_dir = "/home/data/nbc/Laird_PhysicsLearning/dset-v2.0.0/"
subjects = [file.split('/')[-1] for file in sorted(glob.glob(op.join(dset_dir, "sub-*")))]
subjects
```

```python
#Check files
for sub in subjects:
     if len(glob.glob(op.join(dset_dir, sub,'ses-*/func/*fci*.tsv'))) != 12:
        continue
#Ensuring that each subjects IDConn event files aren't empty and have the correct column names
     for IDConn_file in sorted(glob.glob(op.join(dset_dir, sub,'ses-*/func/*fci_run-[1-3]_events.tsv'))):
        df = pd.read_table(IDConn_file)
        if (len(df) != 18):
            print("WARNING, for",sub, 'rows = ', len(df))
        if list(df.columns[:]) != ['onset', 'duration', 'trial_type']:
            print("WARNING, for",sub,'column names are :', list(df.columns[:]))
#Ensuring that each subjects official event files aren't empty and have the correct column names
     for event_file in sorted(glob.glob(op.join(dset_dir, sub,'ses-*/func/*fci_run-[1-3]_acq-official_events.tsv'))):
        df = pd.read_table(event_file)
        if (len(df)) != 18:
            print("WARNING, for",sub, 'rows = ', len(df))
        if list(df.columns[:]) != ['onset', 'duration', 'trial_type', 'trial_phase']:
            print("WARNING, for",sub,'column names are :', list(df.columns[:]))
    #Checking if onset and duration for each subjects official and IDConn event files match
     for official_event_file in sorted(glob.glob(op.join(dset_dir, sub,'ses-*/func/*fci_run-[1-3]_acq-official_events.tsv'))):
            official_df = pd.read_table(official_event_file)
            official_data = official_df[["onset", "duration"]].values
            idconn_event_file = official_event_file.replace("_acq-official", "")
            idconn_df = pd.read_table(idconn_event_file)
            idconn_data = idconn_df[["onset", "duration"]].values
            if not np.array_equal(official_data, idconn_data):
                #f' {sub} doesn't work in python 2
                print("WARNING:", sub, "onset and/or duration DO NOT MATCH for" ,official_event_file, idconn_event_file)
                print(official_data)
                print(idconn_data)
                
                
        
        
        
```

```python
#Session 1
#Nore IDConn files are currenty have the extension '_events.tsv'
order = [0,2,4,1,3,5]
for sub in subjects:
    #Get both the IDConn file from dset v2 and reference file from dset v1 (in Katie's folder)
    for IDConn_file, reference in zip(sorted(glob.glob(op.join(dset_dir,sub, 'ses-1/func/*fci_run-[1-3]_events.tsv'))), sorted(glob.glob(op.join('/home/data/nbc/Laird_PhysicsLearning/dset-v1.0.0/derivatives/idconn-preproc',sub, 'ses-pre/func/*fci_run-[0-2]_events.tsv')))):
        #Get every third onset value, which should correspond to the beginning of each question. Sum duration to obtain the duration of the entire trial 
        df = pd.read_table(IDConn_file)
        data = df[['onset', 'duration']].values
        combined_onset_duration = np.vstack((data[::3,0],np.hstack(([np.sum(data[:,1][0:3]),np.sum(data[:,1][3:6]), np.sum(data[:,1][6:9]), np.sum(data[:,1][9:12]),np.sum(data[:,1][12:15]), np.sum(data[:,1][15:19])])))).T
        IDConn_combined_onset_duration_ordered = np.vstack(([combined_onset_duration[i] for i in order]))
        #official files are the actual events files in dset 2 but have the extension  '_acq-official_events'
        official_file = IDConn_file.replace('_events', '_acq-official_events')
        df = pd.read_table(official_file)
        data = df[['onset', 'duration']].values
        combined_onset_duration = np.vstack((data[::3,0],np.hstack(([np.sum(data[:,1][0:3]),np.sum(data[:,1][3:6]), np.sum(data[:,1][6:9]), np.sum(data[:,1][9:12]),np.sum(data[:,1][12:15]), np.sum(data[:,1][15:19])])))).T
        official_combined_onset_duration_ordered = np.vstack(([combined_onset_duration[i] for i in order]))
        #reference file
        df = pd.read_table(reference)
        reference_file = df[['onset', 'duration']].values                        
        #If not statements to see if onset and duration are equal
        #Comparing official files to reference
        if not np.allclose(official_combined_onset_duration_ordered, reference_file):
            print("WARNING:", sub, "onset and/or duration DO NOT MATCH for" ,reference, official_file)
            print(official_combined_onset_duration_ordered)
            print(reference_file)
        #Else statement to see if the if statement actually works and isn't being skipped over
        else:
            print('official files fine for', sub)
        # Comparing IDConn files to reference
        if not np.allclose(IDConn_combined_onset_duration_ordered, reference_file):
            print("WARNING:", sub, "onset and/or duration DO NOT MATCH for" ,reference, IDConn_file)
            print(reference_file)
            print(IDConn_combined_onset_duration_ordered)
        else:
            print('IDConn files fine for', sub)
                        
```
